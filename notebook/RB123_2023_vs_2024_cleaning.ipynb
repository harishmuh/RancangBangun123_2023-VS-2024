### **Business overview of RancangBangun123**

RancangBangun123 or RB123 is architectural and construction firm that provides end-to-end solutions for residential projects. The company specializes in designing and building high-quality, aesthetically pleasing, and functional structures tailored to meet the specific needs and preferences of its clients. The company's main office is located in Bandung, but they also serve clients from Jakarta, Depok, Tangerang, Bekasi, and surrounding areas.

### **Business Key/Services**

RB123 operates on a full-service business model that encompasses various stages of construction projects, from initial design to final construction and post-construction services. Here are the key aspects of their business:

**1. Architectural Design**

* The company has a team of experienced architects who create custom designs for clients. These designs include detailed blueprints, 3D models, and visualizations to help clients envision their projects.

**2. Construction Services**

* RancangBangun123 offers comprehensive construction services, including site preparation, foundation work, structural construction, finishing, and landscaping. They work with trusted contractors and suppliers to ensure high-quality materials and workmanship.

**3. Project Management**

* The company provides project management services to oversee every aspect of the construction process. This includes scheduling, budgeting, procurement, and coordination with various stakeholders to ensure timely and efficient project completion.

**4. Renovation and Remodeling**

* In addition to new construction, RB123 also specializes in renovation and remodeling projects. They offer services to update and enhance existing structures, including interior redesign, extension, and modernization.

**5. Consultation and Planning**

* The company offers consultation services to help clients plan their construction projects effectively. This includes feasibility studies, cost estimation, regulatory compliance (Assist in obtaining the Building Establishment License (IMB)), and sustainability planning.

## **CONTEXT - CRM ANALYTICS**
CRM analytics involves gathering, analyzing, and interpreting customer data within a Customer Relationship Management (CRM) system to make informed business decisions. Leveraging CRM analytics is crucial for extracting meaningful insights from customer interactions and requirements. These insights can help organizations improve their relationships with customers, enhance customer satisfaction, and drive business growth. By using a CRM system, RB123 can build its customer journey pipeline, tracking progress from initial leads to successful deals (closed wins).

### **Customer Journey of RB123**
The customer journey pipeline in RB123 can be divided into six stages:

**Incoming Leads/Non-Leads**

* These are potential contacts originating from non-customers, such as suppliers (e.g., material suppliers, hosting providers) or other parties looking to collaborate with RB123 (e.g., job seekers).

**Leads**

* Potential leads who have contacted RB123 for consultation or product inquiries. These individuals or families might become future customers seeking construction or renovation services from RB123.

**Qualified Leads**

* Leads who have shown interest in RB123's services have been introduced to RB123's values, products, and services, and have completed the construction/renovation form. They have also agreed to the minimum budget for the project.

**Surveyed Leads or 'Sudah Survei Impresi Leads'**

* Qualified leads who have conducted a detailed meeting with RB123 stakeholders to communicate their project requirements. This meeting can be offline or online, involving the planning team or manager. If conducted on-site, RB employees will typically survey and conduct detailed measurement for the construction area. After this meeting, RB123 staffs will also provide some alternative services and budgets based on the real measurement.

**Close-Win**

* Qualified Leads who have agreed to RB123's terms and have officially become customers by signing the contract.

**Close-Lose**

* Leads who have decided not to proceed with RB123 for their construction/renovation project. Reasons for this could include:
    * **Close-lose (lokasi ga sesuai)**: Unqualified locations (geographically or areas inaccessible by cars)
    * **Close-lose (dibawah minimum budget)**: Inability to meet the minimum budget required by RB123.
    * **Close-lose (waktu butuh lebih cepet)**: Dissatisfaction with RB123's project timeline.
    * **Close-lose (harga kita kemahalan)**: Disagreement with RB123's construction/planning costs despite meeting the minimum budget.
    * **Close-lose (kontrak ga cocok)**: Disagreement with the construction contract terms.
    * **Close-lose (ga diketahui pasti alasannya)**: Cancelling the project without defined reasons.

Based on the visual observation, the numerical data contains outliers

```python
# Calculating outliers and their percentage in features
def calculate_outliers(df_crm, column):
    Q1 = df_crm[column].quantile(0.25)
    Q3 = df_crm[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_crm[(df_crm[column] < lower_bound) | (df_crm[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outlier_data = []

numerical = [var for var in df_crm.columns if df_crm[var].dtypes in ['int64', 'float64'] and var != 'ID']

for column in numerical:
    outlier_result, lower_bound, upper_bound = calculate_outliers(df_crm, column)
    total_outlier = len(outlier_result)
    outlier_percentage = round(total_outlier / len(df_crm[column]) * 100, 2)
    outlier_data.append([column, total_outlier, outlier_percentage, lower_bound, upper_bound])

outlier_df = pd.DataFrame(outlier_data, columns=["Column", "Total Outliers", "Percentage (%)", "Lower Bound", "Upper Bound"])
outlier_df
```

### **Reformating the Lead stage columns**

We will make the new columns of 'Lead stages', instead of using 'Lead stage' as the current 'Lead stage' contains redudant step of 'Close-lose' categories. So we will use only 'Close lose' than using the 'Close-lose (lokasi ga sesuai)','Close-lose (dibawah minimum budget)','Close-lose (kontrak ga cocok)','Close-lose (waktu butuh lebih cepet)','Close-lose (harga kita kemahalan)','Close-lose (ga diketahui pasti alasannya)'

```python
# Modifying and re-defining 'lead stage' column into 'lead stages'
# Modification will be based on the 7 stages of leading and reducing the redundancy of category in 'lead stage'

# Define the function to apply
def categorize_lead_stage(stage):
    close_lose_conditions = [
        'Close-lose (lokasi ga sesuai)',
        'Close-lose (dibawah minimum budget)',
        'Close-lose (kontrak ga cocok)',
        'Close-lose (waktu butuh lebih cepet)',
        'Close-lose (harga kita kemahalan)',
        'Close-lose (ga diketahui pasti alasannya)'
    ]
    if stage in close_lose_conditions:
        return 'Close-lose'
    return stage

# Apply the function to create the new column
df_crm['Lead stages'] = df_crm['Lead stage'].apply(categorize_lead_stage)

# Display the DataFrame to verify the new column
display(df_crm.sample(5))
```

**Handling missing values**

In this step, we want to minimize data loss as much as possible for the analysis. For handling missing values, we will analyze the reason why the data are missing while reviewing their quantity, distribution, and correlation. After that, we will decide each case separately. As there might be some columns that contains missing values and we have no sufficient information to replace those missing values, so will let them as they are or We will handle them later just before the data analysis process.



For columns with string or object data type:
* **Marketing Channel** and **Marketing Channel description** have a small percentage of missing values (less than 5%) and contain texts or object data type. The reason for the missing probably due to errors in the data input or data collection process. These missing values will be handled using an imputation method with 'Not answering'
* **Customer message** contains text or object data type and has 35% of missing values. The reason of the missing probably may be due to errors in the data input or the data collection process. However, there is also probability is because the customer does not provide sufficient information (their motive for contacting RB123), so we cannot record the text. The missing values of this column will be filled with 'NA'
* **Furniture order** contains text or object data type. This column has almost 80% of missing values. The reason behind is more likely due to many leads do not provide an answer during the WhatsApp interview or do not want include additional custom furniture for a construction/renovation project. Because of that, we will fill the column with 'No furniture'  

For columns with numeric or float data type:
* **KM**, **KT**, **LB**, **LT**, all of them have more than 60% of missing values. Based on the domain knowledge and the present information, we cannot figure out the way to fill in the missing values. So, we will exclude them fromthe  analysis


```python
# Replace missing values in 'Marketing channel' and 'Marketing channel description' with 'Not answering'
df_crm['Marketing channel'] = df_crm['Marketing channel'].fillna('Not answering')
df_crm['Marketing channel description'] = df_crm['Marketing channel description'].fillna('Not answering')
```

```python
# Replacing missing values in 'Customer message' with 'NA'
df_crm['Customer message'] = df_crm['Customer message'].fillna('NA')
```

```python
# Replacing missing values in 'Furniture order' with 'No furniture'
df_crm['Furniture order'] = df_crm['Furniture order'].fillna('No furniture')
```
